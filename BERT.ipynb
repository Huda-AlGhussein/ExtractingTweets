```python
import numpy as np
import pandas as pd
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
import requests
from bs4 import BeautifulSoup
import re
```

```python
tokenizer = AutoTokenizer.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')

model = AutoModelForSequenceClassification.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')
```

```python
sentences= ['It was good but couldve been better. Great', 'Meh, it was ok', 'it was below average','This was the worst', 'Amazing loved it best I have ever seen']
tokens= []
for sen in sentences:
  t = tokenizer.encode(sen, return_tensors='pt')
  tokens.append(t)
print(tokens)
```

```python
for t in tokens:
  result = model(t)
  print(int(torch.argmax(result.logits))+1)
```
